This notebook visualizes the structure and coverage of the Kitchen Safety and Sanitation
Standard Operating Procedures authored by Keandra Morris for Airstreams Renewables, Inc.

The SOP follows the HACCP (Hazard Analysis Critical Control Point) framework and covers
6 major compliance areas across 30+ documented procedures.

**SOP Sections:**
1. HACCP Principles (7 principles)
2. Health & Hygiene
3. Food Handling
4. Cleaning & Sanitizing
5. First Aid & Emergencies
6. COVID-19 Procedures

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import warnings
warnings.filterwarnings('ignore')

GREEN = '#1B4332'
GOLD  = '#B8933A'
MUTED = '#4A5568'
LGRAY = '#9CA3AF'
RED   = '#C0392B'
BG    = '#F7F6F1'
WHITE = '#FFFFFF'

print('Libraries loaded ✓')
```

## 1. SOP Coverage Summary

```python
sop_sections = {
    'HACCP Principles':        {'procedures': 7,  'color': GREEN,     'icon': '01'},
    'Health & Hygiene':        {'procedures': 5,  'color': '#2C5282', 'icon': '02'},
    'Food Handling':           {'procedures': 11, 'color': '#7B3F00', 'icon': '03'},
    'Cleaning & Sanitizing':   {'procedures': 5,  'color': '#1A5276', 'icon': '04'},
    'First Aid & Emergencies': {'procedures': 6,  'color': RED,       'icon': '05'},
    'COVID-19 Procedures':     {'procedures': 4,  'color': MUTED,     'icon': '06'},
}

total_procedures = sum(v['procedures'] for v in sop_sections.values())

print('SOP COVERAGE SUMMARY')
print('─' * 40)
for section, data in sop_sections.items():
    print(f"  {data['icon']}. {section:<28} {data['procedures']} procedures")
print('─' * 40)
print(f'  Total documented procedures:  {total_procedures}')
print(f'  Compliance areas covered:     {len(sop_sections)}')
print(f'  HACCP principles applied:     7')
print(f'  Supporting log types:         4')
print(f'  Audit findings (critical):    0')
```

## 2. Compliance Framework Visualization

```python
section_details = [
    ('01', 'HACCP Principles', GREEN,
     ['Hazard Analysis', 'Critical Control Points (CCPs)', 'Critical Limits',
      'CCP Monitoring', 'Corrective Action', 'Verification', 'Recordkeeping']),
    ('02', 'Health & Hygiene', '#2C5282',
     ['Personal grooming standards', 'Dress code & PPE requirements',
      'Handwashing protocol (6-step)', 'Glove use procedures',
      'No food/drink/smoking policy']),
    ('03', 'Food Handling', '#7B3F00',
     ['Purchasing standards', 'Receiving & inspection logs',
      'Cold/dry storage controls', 'Thaw/cook/cool/reheat temps',
      'Temperature danger zone (41-135F)', 'FIFO & labeling',
      'Allergen management protocol']),
    ('04', 'Cleaning & Sanitizing', '#1A5276',
     ['Dishwashing procedures', 'Disinfecting & sanitizing steps',
      'Chemical storage requirements', 'Deep cleaning schedule',
      'Sanitation logs & recordkeeping']),
    ('05', 'First Aid & Emergencies', RED,
     ['Slip & fall response protocol', 'First aid kit & procedures',
      'Cut injury & blood exposure', 'Grease fire response',
      'Pest control procedures', 'Allergic reaction safety plan']),
    ('06', 'COVID-19 Procedures', MUTED,
     ['Kitchen PPE requirements', 'Social distancing protocols',
      'Cafeteria layout & table distancing',
      'Surface disinfecting schedule']),
]

fig = plt.figure(figsize=(14, 9), facecolor=BG)

# Header
hdr = fig.add_axes([0, 0.91, 1, 0.09])
hdr.set_facecolor(GREEN)
hdr.text(0.5, 0.65, 'Kitchen Safety & Sanitation — SOP Compliance Framework',
         ha='center', va='center', color='white', fontsize=14, fontweight='bold')
hdr.text(0.5, 0.18,
         'HACCP-Based Policy  ·  Airstreams Renewables, Inc.  ·  Authored by Keandra Morris',
         ha='center', va='center', color=GOLD, fontsize=9)
hdr.set_xticks([]); hdr.set_yticks([])
for sp in hdr.spines.values(): sp.set_visible(False)

# Summary metric cards
metrics = [
    ('6',    'Policy\nSections'),
    ('7',    'HACCP\nPrinciples'),
    ('30+',  'Documented\nProcedures'),
    ('4',    'Compliance\nLog Types'),
    ('100%', 'Audit\nReadiness'),
    ('0',    'Critical\nFindings'),
]
for i, (val, label) in enumerate(metrics):
    card = fig.add_axes([0.01 + i*0.164, 0.79, 0.155, 0.11])
    card.set_facecolor(WHITE)
    card.set_xticks([]); card.set_yticks([])
    for sp in card.spines.values(): sp.set_color('#D1D5DB')
    color = GOLD if i == 5 else GREEN
    card.text(0.5, 0.72, val, ha='center', va='center', fontsize=20,
              color=color, fontweight='bold', transform=card.transAxes)
    card.text(0.5, 0.22, label, ha='center', va='center', fontsize=7.5,
              color=MUTED, transform=card.transAxes, linespacing=1.4)

# Section blocks (2 rows x 3 cols)
positions = [(0.01,0.31),(0.34,0.31),(0.67,0.31),(0.01,0.06),(0.34,0.06),(0.67,0.06)]

for (x, y), (num, title, color, items) in zip(positions, section_details):
    ax_s = fig.add_axes([x, y, 0.31, 0.23])
    ax_s.set_facecolor(WHITE)
    ax_s.set_xticks([]); ax_s.set_yticks([])
    for sp in ax_s.spines.values(): sp.set_color('#E5E7EB')

    # Color left accent bar
    bar = fig.add_axes([x, y, 0.007, 0.23])
    bar.set_facecolor(color)
    bar.set_xticks([]); bar.set_yticks([])
    for sp in bar.spines.values(): sp.set_visible(False)

    ax_s.text(0.05, 0.93, f'Section {num}', ha='left', va='top', fontsize=7.5,
              color=color, fontweight='bold', transform=ax_s.transAxes)
    ax_s.text(0.05, 0.83, title, ha='left', va='top', fontsize=10,
              color='#1C1C1C', fontweight='bold', transform=ax_s.transAxes)

    y_step = 0.60 / len(items)
    for j, item in enumerate(items):
        yp = 0.70 - j * y_step
        ax_s.text(0.055, yp, f'• {item}', ha='left', va='top', fontsize=7.5,
                  color=MUTED, transform=ax_s.transAxes)

plt.figtext(0.5, 0.005,
    'Keandra Morris  ·  Operations Project Manager  ·  keandrajm.github.io/PM_Portfolio/',
    ha='center', fontsize=7.5, color=LGRAY, style='italic')

plt.show()
```

## 3. Procedure Count by Section

```python
fig, ax = plt.subplots(figsize=(10, 4), facecolor=BG)
ax.set_facecolor(WHITE)

labels = list(sop_sections.keys())
counts = [v['procedures'] for v in sop_sections.values()]
colors = [v['color'] for v in sop_sections.values()]

bars = ax.barh(labels, counts, color=colors, alpha=0.82, height=0.55, edgecolor='none')

for bar, count in zip(bars, counts):
    ax.text(bar.get_width() + 0.15, bar.get_y() + bar.get_height()/2,
            f'{count} procedures', va='center', fontsize=9, color=MUTED)

ax.set_xlabel('Number of Documented Procedures', fontsize=9)
ax.set_title(f'SOP Coverage by Section  (Total: {total_procedures} procedures)',
             fontsize=11, fontweight='bold', color=GREEN, pad=10)
ax.set_xlim(0, 15)
ax.grid(axis='x', alpha=0.3, linewidth=0.7)
ax.tick_params(labelsize=9)
for sp in ['top','right']: ax.spines[sp].set_visible(False)

plt.tight_layout()
plt.show()
```
