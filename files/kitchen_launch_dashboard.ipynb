This notebook builds a Power BI-style project launch dashboard tracking all 15 tasks
completed across the 120-day Airstreams Renewables cafeteria program build-out.

**Dashboard panels:**
- Six executive KPI cards (Total Tasks, Completed, In Progress, Pending, Project Days, On-Time Rate)
- Project task timeline (Gantt) with priority color coding and phase bands
- Status breakdown (donut)
- Priority distribution (donut)
- Tasks by category (horizontal bar)
- Full task detail table

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import FancyBboxPatch, Wedge
import numpy as np
import warnings
warnings.filterwarnings('ignore')

GREEN  = '#1B4332'
GOLD   = '#B8933A'
DARK   = '#1C1C1C'
MUTED  = '#4A5568'
LGRAY  = '#9CA3AF'
BG     = '#F0F0EC'
WHITE  = '#FFFFFF'
RED    = '#C0392B'
TEAL   = '#2980B9'
LIGHT  = '#E8F4F0'

print('Libraries loaded ✓')
```

## 1. Define Launch Tasks

```python
# (Name, Priority, Category, Start Day, Duration Days, % Complete)
tasks = [
    ('SOP Scope & Objectives',        'High',   'SOPs',         1,  14, 100),
    ('Kitchen Equipment Order',        'High',   'Equipment',    5,  30, 100),
    ('Vendor Agreements',              'High',   'Vendors',      8,  20, 100),
    ('Furniture & Supplies',           'Medium', 'Equipment',   15,  28, 100),
    ('Hiring Process',                 'High',   'Staffing',    20,  25, 100),
    ('Menu & Supplier Setup',          'Medium', 'Vendors',     28,  22, 100),
    ('Training Requirements Design',   'Medium', 'Training',    35,  20, 100),
    ('Uniform & Safety Supplies',      'Low',    'Equipment',   38,  15, 100),
    ('Workspace Materials Setup',      'Medium', 'Equipment',   40,  18, 100),
    ('SOP Documentation & Review',     'High',   'SOPs',        42,  25, 100),
    ('POS System Configuration',       'High',   'Technology',  52,  20, 100),
    ('Staff Training Delivery',        'Medium', 'Training',    60,  22, 100),
    ('Pre-Launch Inspection',          'High',   'Compliance',  78,  12, 100),
    ('Soft Launch & Issue Resolution', 'High',   'Launch',      88,  15, 100),
    ('Full Kitchen Operations Launch', 'High',   'Launch',     108,  12, 100),
]

priority_colors = {'High': GREEN, 'Medium': GOLD, 'Low': LGRAY}
category_colors = {
    'SOPs':       '#1B4332',
    'Equipment':  '#2980B9',
    'Vendors':    '#B8933A',
    'Staffing':   '#7B3F00',
    'Training':   '#6C3483',
    'Technology': '#117A65',
    'Compliance': '#C0392B',
    'Launch':     '#4A5568',
}

high_count = sum(1 for t in tasks if t[1]=='High')
med_count  = sum(1 for t in tasks if t[1]=='Medium')
low_count  = sum(1 for t in tasks if t[1]=='Low')

print('LAUNCH TASK SUMMARY')
print('─' * 35)
print(f'  Total tasks:        {len(tasks)}')
print(f'  Completed:          {len(tasks)} (100%)')
print(f'  High priority:      {high_count}')
print(f'  Medium priority:    {med_count}')
print(f'  Low priority:       {low_count}')
print(f'  Project duration:   120 days')
print(f'  On-time rate:       100%')
```

## 2. Full Launch Dashboard

```python
fig = plt.figure(figsize=(15, 10), facecolor=BG)

# Header
hdr = fig.add_axes([0, 0.935, 1, 0.065])
hdr.set_facecolor(GREEN)
hdr.text(0.012, 0.65, 'KITCHEN OPERATIONS LAUNCH DASHBOARD',
         ha='left', va='center', color=WHITE, fontsize=14, fontweight='bold')
hdr.text(0.012, 0.18, 'Project Overview  ·  Airstreams Renewables, Inc.  ·  All 15 Tasks Completed On Time',
         ha='left', va='center', color=GOLD, fontsize=8.5)
hdr.text(0.988, 0.42, '120-Day Program  ·  Authored by Keandra Morris',
         ha='right', va='center', color=WHITE, fontsize=8, style='italic')
hdr.set_xticks([]); hdr.set_yticks([])
for sp in hdr.spines.values(): sp.set_visible(False)

# KPI Cards
kpi_data = [
    ('Total Tasks',  '15',   'Launch tasks tracked',  GREEN),
    ('Completed',    '15',   'All tasks finished',    '#117A65'),
    ('In Progress',  '0',    'At go-live date',       GOLD),
    ('Pending',      '0',    'Outstanding items',     LGRAY),
    ('Project Days', '120',  'End-to-end duration',   TEAL),
    ('On-Time Rate', '100%', 'Zero overdue tasks',    '#6C3483'),
]
for i, (title, val, sub, color) in enumerate(kpi_data):
    ax_c = fig.add_axes([0.008 + i*0.164, 0.83, 0.155, 0.095])
    ax_c.set_facecolor(WHITE)
    ax_c.set_xticks([]); ax_c.set_yticks([])
    for sp in ax_c.spines.values(): sp.set_color('#D1D5DB')
    ax_c.axvline(x=0, ymin=0.1, ymax=0.9, linewidth=4, color=color)
    ax_c.text(0.13, 0.76, title, transform=ax_c.transAxes, fontsize=7.5, color=MUTED)
    ax_c.text(0.13, 0.36, val,   transform=ax_c.transAxes, fontsize=22,  color=color, fontweight='bold')
    ax_c.text(0.13, 0.07, sub,   transform=ax_c.transAxes, fontsize=7,   color=LGRAY)

# Gantt
ax_g = fig.add_axes([0.008, 0.32, 0.58, 0.49], facecolor=WHITE)
for sp in ax_g.spines.values(): sp.set_color('#E5E7EB')
for i, (name, pri, cat, start, dur, pct) in enumerate(tasks):
    ax_g.barh(i, 120, left=0, height=0.6, color='#E8ECE8', edgecolor='none')
    ax_g.barh(i, dur, left=start, height=0.6, color=priority_colors[pri], alpha=0.85, edgecolor='none')
    ax_g.text(start + dur + 1.5, i, '✓', va='center', fontsize=7, color=priority_colors[pri], fontweight='bold')

ax_g.axvspan(0,  50,  alpha=0.04, color=TEAL,  zorder=0)
ax_g.axvspan(50, 90,  alpha=0.04, color=GOLD,  zorder=0)
ax_g.axvspan(90, 120, alpha=0.04, color=GREEN, zorder=0)
ax_g.text(25,  14.7, 'Phase 1: Build',    ha='center', fontsize=7, color=TEAL,  fontweight='bold')
ax_g.text(70,  14.7, 'Phase 2: Prepare',  ha='center', fontsize=7, color=GOLD,  fontweight='bold')
ax_g.text(105, 14.7, 'Phase 3: Launch',   ha='center', fontsize=7, color=GREEN, fontweight='bold')
ax_g.axvline(120, color=GREEN, linestyle='--', linewidth=1.2, alpha=0.8)

ax_g.set_yticks(range(len(tasks)))
ax_g.set_yticklabels([t[0] for t in tasks], fontsize=7.8)
ax_g.set_xlabel('Project Day', fontsize=8, color=MUTED)
ax_g.set_xlim(-2, 126); ax_g.set_ylim(-0.7, 15.3)
ax_g.invert_yaxis()
ax_g.set_title('Project Task Timeline', fontsize=10, fontweight='bold', color=DARK, loc='left', pad=7)
ax_g.grid(axis='x', alpha=0.25, linewidth=0.6)
ax_g.tick_params(labelsize=7.8)
for sp in ['top','right']: ax_g.spines[sp].set_visible(False)
legend_patches = [mpatches.Patch(color=c, label=p, alpha=0.85) for p, c in priority_colors.items()]
ax_g.legend(handles=legend_patches, title='Priority', title_fontsize=7.5,
            fontsize=7.5, frameon=True, framealpha=0.9, edgecolor='#D1D5DB',
            loc='lower right', ncol=3)

# Status Breakdown
ax_s = fig.add_axes([0.608, 0.61, 0.19, 0.20], facecolor=WHITE)
for sp in ax_s.spines.values(): sp.set_color('#E5E7EB')
ax_s.set_xticks([]); ax_s.set_yticks([])
ax_s.set_title('Status Breakdown', fontsize=9, fontweight='bold', color=DARK, loc='left', pad=5)
w = Wedge((0.5, 0.44), 0.35, 0, 360, width=0.14, facecolor='#117A65', edgecolor=WHITE, linewidth=1.5)
ax_s.add_patch(w)
ax_s.text(0.5, 0.50, '15',          ha='center', va='center', fontsize=22, color=GREEN,    fontweight='bold', transform=ax_s.transAxes)
ax_s.text(0.5, 0.30, 'Completed',   ha='center', va='center', fontsize=8,  color=MUTED,    transform=ax_s.transAxes)
ax_s.text(0.5, 0.10, '100% On Time',ha='center', va='center', fontsize=7.5,color='#117A65',fontweight='bold', transform=ax_s.transAxes)
ax_s.set_xlim(0,1); ax_s.set_ylim(0,1)

# Priority Breakdown
ax_p = fig.add_axes([0.608, 0.38, 0.19, 0.20], facecolor=WHITE)
for sp in ax_p.spines.values(): sp.set_color('#E5E7EB')
ax_p.set_xticks([]); ax_p.set_yticks([])
ax_p.set_title('Priority Breakdown', fontsize=9, fontweight='bold', color=DARK, loc='left', pad=5)
start_a = 90
for size, color in zip([high_count, med_count, low_count], [GREEN, GOLD, LGRAY]):
    angle = 360 * size / 15
    Wedge_p = Wedge((0.5,0.47), 0.32, start_a-angle, start_a, width=0.35, facecolor=color, edgecolor=WHITE, linewidth=1.5)
    ax_p.add_patch(Wedge_p)
    start_a -= angle
ax_p.text(0.18, 0.52, f'High:   {high_count}', fontsize=8, color=GREEN, fontweight='bold', transform=ax_p.transAxes)
ax_p.text(0.18, 0.34, f'Medium: {med_count}',  fontsize=8, color=GOLD,  fontweight='bold', transform=ax_p.transAxes)
ax_p.text(0.18, 0.16, f'Low:    {low_count}',   fontsize=8, color=LGRAY, fontweight='bold', transform=ax_p.transAxes)
ax_p.set_xlim(0,1); ax_p.set_ylim(0,1)

# Tasks by Category
cat_counts = {}
for t in tasks:
    cat_counts[t[2]] = cat_counts.get(t[2], 0) + 1
cat_labels = list(cat_counts.keys())
cat_vals   = list(cat_counts.values())
cat_clrs   = [category_colors[c] for c in cat_labels]

n_cats = len(cat_labels)
ax_cat = fig.add_axes([0.608, 0.32, 0.38, 0.045 + 0.038*n_cats], facecolor=WHITE)
for sp in ax_cat.spines.values(): sp.set_color('#E5E7EB')
ax_cat.barh(range(n_cats), cat_vals, color=cat_clrs, alpha=0.85, height=0.6, edgecolor='none')
for yi, val in enumerate(cat_vals):
    ax_cat.text(val + 0.05, yi, str(val), va='center', fontsize=8, color=MUTED)
ax_cat.set_yticks(range(n_cats))
ax_cat.set_yticklabels(cat_labels, fontsize=8)
ax_cat.set_xlim(0, max(cat_vals) + 0.8)
ax_cat.set_title('Tasks by Category', fontsize=9, fontweight='bold', color=DARK, loc='left', pad=5)
ax_cat.grid(axis='x', alpha=0.2, linewidth=0.6)
for sp in ['top','right']: ax_cat.spines[sp].set_visible(False)

# Task Detail Table
ax_tbl = fig.add_axes([0.008, 0.04, 0.975, 0.265], facecolor=WHITE)
for sp in ax_tbl.spines.values(): sp.set_color('#E5E7EB')
ax_tbl.set_xticks([]); ax_tbl.set_yticks([])
ax_tbl.set_title('Task Detail', fontsize=9, fontweight='bold', color=DARK, loc='left', pad=5)

col_x = [0.01, 0.04, 0.46, 0.555, 0.625, 0.695, 0.765, 0.855]
col_headers = ['#', 'Task Name', 'Priority', 'Start Day', 'End Day', 'Duration', 'Status', 'Category']
row_h = 0.88 / (len(tasks) + 1.5)

hdr_bg = FancyBboxPatch((0,0.87), 1, 0.13, transform=ax_tbl.transAxes,
                         boxstyle='round,pad=0', facecolor=GREEN, edgecolor='none')
ax_tbl.add_patch(hdr_bg)
for hdr_txt, hx in zip(col_headers, col_x):
    ax_tbl.text(hx, 0.925, hdr_txt, transform=ax_tbl.transAxes,
                fontsize=7.5, color=WHITE, fontweight='bold', va='center')

for i, (name, pri, cat, start, dur, pct) in enumerate(tasks):
    y = 0.85 - (i+0.5) * row_h
    bg = LIGHT if i % 2 == 0 else WHITE
    ax_tbl.add_patch(FancyBboxPatch((0, y-row_h*0.45), 1, row_h*0.9,
                     transform=ax_tbl.transAxes, boxstyle='round,pad=0',
                     facecolor=bg, edgecolor='none'))
    pri_c = priority_colors[pri]
    for rx, txt, clr, bold in [
        (col_x[0], f'{i+1}',         MUTED,                False),
        (col_x[1], name,             DARK,                 False),
        (col_x[2], pri,              pri_c,                True),
        (col_x[3], str(start),       MUTED,                False),
        (col_x[4], str(start+dur),   MUTED,                False),
        (col_x[5], f'{dur}d',        MUTED,                False),
        (col_x[6], '✓ Done',         '#117A65',            True),
        (col_x[7], cat,              category_colors[cat], False),
    ]:
        ax_tbl.text(rx, y, txt, transform=ax_tbl.transAxes, fontsize=7,
                    color=clr, va='center', fontweight='bold' if bold else 'normal')

ax_tbl.set_xlim(0,1); ax_tbl.set_ylim(0,1)

plt.figtext(0.5, 0.005,
    'Keandra Morris  ·  Operations Project Manager  ·  keandrajm.github.io/PM_Portfolio/',
    ha='center', fontsize=7.5, color=LGRAY, style='italic')

plt.show()
```
