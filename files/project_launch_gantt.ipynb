This notebook visualizes the phased 120-day launch timeline for the Airstreams Renewables
cafeteria and dining program — built from scratch with no prior infrastructure.

**Timeline covers:**
- Phase 1: Build (Days 1–50): Scoping, vendor selection, equipment, layout
- Phase 2: Prepare (Days 51–90): Hiring, SOP documentation, training, systems setup, inspections
- Phase 3: Launch (Days 91–120): Soft launch, issue resolution, full operations go-live
- Key milestones: Scope Approval, Vendor Contracts, Training Complete, GO LIVE

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import warnings
warnings.filterwarnings('ignore')

GREEN = '#1B4332'
GOLD  = '#B8933A'
MUTED = '#4A5568'
LGRAY = '#9CA3AF'
RED   = '#C0392B'
TEAL  = '#2C7BB6'
BG    = '#F7F6F1'

print('Libraries loaded ✓')
```

## 1. Define Launch Tasks

```python
# (Task name, start day, duration days, category, color)
tasks = [
    ('Program Scoping & Requirements',       1,  18, 'Planning',    TEAL),
    ('Vendor Selection & Contracts',         10, 25, 'Procurement', GOLD),
    ('Equipment Procurement & Install',      28, 40, 'Procurement', GOLD),
    ('Layout Design & Build-out',            22, 50, 'Planning',    TEAL),
    ('Staff Recruiting & Hiring',            30, 22, 'People',      '#7B3F00'),
    ('SOP & Compliance Documentation',       35, 35, 'Systems',     GREEN),
    ('Staff Training Program Development',   40, 25, 'People',      '#7B3F00'),
    ('Technology Setup (Planner, POS, iPad)',52, 28, 'Systems',     GREEN),
    ('Health & Safety Inspections',          62, 20, 'Compliance',  RED),
    ('Staff Training Delivery',              65, 22, 'People',      '#7B3F00'),
    ('Dry Run / Soft Launch',                85, 15, 'Launch',      MUTED),
    ('Issue Resolution & Final Prep',        95, 15, 'Launch',      MUTED),
    ('Full Operations Launch',              110, 10, 'Launch',      GREEN),
]

categories = {'Planning': TEAL, 'Procurement': GOLD, 'People': '#7B3F00',
               'Systems': GREEN, 'Compliance': RED, 'Launch': MUTED}

print(f'Total tasks tracked: {len(tasks)}')
print(f'Categories: {list(categories.keys())}')
```

## 2. Build the Gantt Chart

```python
fig, ax = plt.subplots(figsize=(14, 8), facecolor=BG)
ax.set_facecolor('#FFFFFF')

y_labels = [t[0] for t in tasks]

for i, (name, start, dur, cat, color) in enumerate(tasks):
    ax.barh(i, dur, left=start, height=0.55, color=color, alpha=0.82, edgecolor='none')
    ax.text(start + dur + 1, i, f'Day {start+dur}', va='center', fontsize=7.5, color=MUTED)

# Milestone markers
milestones = [
    (18,  'Scope Approved'),
    (55,  'Vendors Contracted'),
    (88,  'Training Complete'),
    (110, 'GO LIVE'),
]
for day, label in milestones:
    ax.plot(day, -0.8, 'D', color=GOLD, markersize=9, zorder=5)
    ax.text(day, -1.4, label, ha='center', fontsize=7, color=GOLD, fontweight='bold')

# Phase bands
phase_bands = [
    (1,  50,  TEAL,  'Phase 1: Build'),
    (51, 90,  GOLD,  'Phase 2: Prepare'),
    (91, 120, GREEN, 'Phase 3: Launch'),
]
for s, e, c, label in phase_bands:
    ax.axvspan(s, e, alpha=0.04, color=c)
    ax.text((s+e)/2, -0.15, label, ha='center', fontsize=7.5, color=c,
            fontweight='bold', transform=ax.get_xaxis_transform())

ax.axvline(110, color=GREEN, linestyle='--', linewidth=1.2, alpha=0.6)
ax.text(111, len(tasks)-0.8, 'Launch', fontsize=8, color=GREEN, fontweight='bold')

ax.set_yticks(range(len(tasks)))
ax.set_yticklabels(y_labels, fontsize=8.5)
ax.set_xlabel('Project Day', fontsize=9)
ax.set_title('120-Day Cafeteria Program Launch  ·  Airstreams Renewables',
             fontsize=13, fontweight='bold', color=GREEN, pad=12)
ax.set_xlim(0, 128)
ax.set_ylim(-2.5, len(tasks) - 0.3)
ax.invert_yaxis()
ax.grid(axis='x', alpha=0.3, linewidth=0.7)
ax.tick_params(labelsize=8)
for sp in ['top','right']: ax.spines[sp].set_visible(False)

legend_patches = [mpatches.Patch(color=c, label=cat, alpha=0.82) for cat, c in categories.items()]
ax.legend(handles=legend_patches, loc='lower right', fontsize=8, frameon=True,
          framealpha=0.9, edgecolor='#D1D5DB', ncol=3)

plt.figtext(0.5, 0.005,
    'Keandra Morris  ·  Operations Project Manager  ·  keandrajm.github.io/PM_Portfolio/',
    ha='center', fontsize=7.5, color=LGRAY, style='italic')

plt.tight_layout(rect=[0, 0.03, 1, 1])
plt.show()
```
